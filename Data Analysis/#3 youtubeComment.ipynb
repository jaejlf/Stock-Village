```python
from googleapiclient.discovery import build
import pandas as pd
import numpy as np
```

```python
def getComment(keyword,video_id):
    api_key = 'INPUT_YOUR_KEY'
    comments = list()
    api_obj = build('youtube', 'v3', developerKey=api_key)
    #키워드에 대한 내용이 적힌 댓글들만 관련성 있는 댓글로 최대 1개 수집
    response = api_obj.commentThreads().list(part='snippet', videoId=video_id, maxResults=1,order='relevance',searchTerms = keyword).execute()
    while response:
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments.append([comment['textDisplay'],comment['likeCount']])
            #답글이 달려있는 경우 추가로 수집(현재는 수집 X)
            if item['snippet']['totalReplyCount'] < 0:
                for reply_item in item['replies']['comments']:
                    reply = reply_item['snippet']
                    comments.append([reply['textDisplay'],reply['likeCount']])            
        if 'nextPageToken' in response:
            response = api_obj.commentThreads().list(part='snippet', videoId=video_id, pageToken=response['nextPageToken'],maxResults=1,order='relevance',searchTerms = keyword).execute()
        else:
            break
    return comments
```

#1 Video ID 저장된 csv 파일 열기

```python
data = pd.read_csv('네이버 댓글.csv')
data
```

#2 Video ID에 해당하는 댓글 정보를 데이터 프레임에 저장한다.

```python
comment = {'Stock' : [], 'Date' : [], 'Video ID' : [], 'Comment' : [], 'Comment Like': []}
result = pd.DataFrame(comment)

for i in range(len(data)):
    try:
        comments = getComment('네이버',data['Video ID'][i])
        for j in range(len(comments)):
            comment = {'Stock' : [data['Stock'][i]], 'Date' : [data['Date'][i]], 'Video ID' : [data['Video ID'][i]], 'Comment' : [comments[j][0]], 'Comment Like': [comments[j][1]]}
            tmp = pd.DataFrame(comment)
            result = pd.concat([result,tmp])
    except:
        print('ID: '+data['Video ID'][i]+'를 가진 영상이 삭제되었거나 API 할당량이 초과되었습니다.')
result
```

```python
result.to_csv('Comment.csv',index = False)
```
